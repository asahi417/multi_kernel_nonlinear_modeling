# Visualize result of grid search
- ` python Gridsearch_1d.py`
- plot result for each `eta`

```python
% matplotlib inline
from matplotlib import pyplot as plt
```

```python
import numpy as np
from glob import glob
from model import nonlinear_signal_1d as n_sig
from model import MultiKernelRegressionFBS01 as Regressor_fbs
from model import MultiKernelRegressionPDA01 as Regressor_pda
from model.util import plot_full, save_stats

def tmp_plot(target_eta, target_key=None):
    # order 1 for 1d signal and 2 for 2d signal
    order = "1"
    
    dict_size, name, error = [], [], []
    for _path in glob("./results/%sd/*" % order):
        _name = _path.split("/")[-1].replace(".npz", "")
        if "fbs" in _name and "alpha_1e-06" in _name:
            continue
        if _name.split("eta_")[-1] not in target_eta:
            continue
        if target_key:
            if _name in target_key:
                continue
        error.append(np.load(_path)["error"])
        dict_size.append(np.load(_path)["dict_size"])
        name.append(_name)
    plot_full(error, dict_size, name)
```

```python
target_eta = ["0.08", "0.09", "0.1", "0.15", "0.2", "0.25", "0.3"]
```

```python
_target = target_eta[0]
# target_key = ["fbs_dict_1e-10_alpha_1e-06_eta_0.08"]
target_key = None
tmp_plot([_target], target_key=target_key)
```
