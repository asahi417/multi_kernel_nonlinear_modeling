```python
% matplotlib inline
from matplotlib import pyplot as plt
```

```python
import numpy as np
from itertools import product
from model import MultiKernelRegressionFBS as Regressor_fbs
from model import MultiKernelRegressionPDA as Regressor_pda
from model import nonlinear_signal_1d as n_sig
```

```python
def ma(target, window):
    return [np.mean(target[_s:_s+window]) for _s in range(len(target)-window)]


def plot_full(regressors, name, window=50):
    plt.rcParams['font.family'] = 'Times New Roman'
    plt.rcParams['font.size'] = 18

    plt.figure(0)
    for regressor, _name in zip(regressors, name):
        plt.plot(regressor.r_dict_size, label=_name)
    plt.legend(loc="upper left")
    plt.show()
    
    plt.figure(1)
    for regressor, _name in zip(regressors, name):
        plt.plot(ma(regressor.r_error, window), label=_name)
        plt.yscale("log")
    plt.legend(loc="upper right")
    plt.show()
```

```python
size = 7000
noise_std = 0.01
x, y, y_noise = n_sig(size, noise_std)
```

```python
a = np.arange(1, 10)
b = np.arange(-4, 2)

kernel_parameter = [i[0]*10.0**i[1] for i in product(a, b)]
eta0 = 0.3
dict_size = 20
```

```python
dictionary_th = 10**-11
alpha_dict = 10**-7
alpha_element = alpha_dict*10**-1
regressor1 = Regressor_fbs(
    eta0=eta0,
    kernel_parameter=kernel_parameter,
    dictionary_th=dictionary_th,
    dict_size=dict_size,
    alpha_dict=alpha_dict,
    alpha_element=alpha_element
)
regressor1.fit(x, y)

regressor2 = Regressor_pda(
    eta0=eta0,
    kernel_parameter=kernel_parameter,
    dictionary_th=dictionary_th,
    dict_size=dict_size,
    alpha_dict=alpha_dict,
    alpha_element=alpha_element
)

regressor2.fit(x, y)

plot_full([regressor1, regressor2], name=["FBS", "PDA"])
```
